```python
# importing dependencies here
import numpy as np
import pandas as pd
import os
import seaborn as sns
from matplotlib import pyplot as plt


import re
import nltk


from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
mbti_data = pd.read_csv('mbti_1.csv')
mbti_data.head()
```

```python
def unique_words(s):
    unique = set(s.split(" "))
    return len(unique) / 50

def colons(post):
    # Includes colons used in emojis
    colon_count = 0
    words = post.split()
    for e in words:
        if "http" not in e:
            colon_count += e.count(":")
    return colon_count / 50
```

```python
mbti_data['upper']= [len(re.findall('[A-Z]',post)) for post in mbti_data['posts']]
```

```python
mbti_data['symbols']= [len(re.findall('[@ # $ % & . : ;]',post)) for post in mbti_data['posts']]
```
